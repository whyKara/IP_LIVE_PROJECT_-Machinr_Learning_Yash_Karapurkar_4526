```python
import pandas as pd


df = pd.read_csv("Data.csv",index_col=0)
df.head()
```

```python
for i in range(len(df)):
    if df.Timing[i] == 'pm':
        df.LoginTime[i]=df.LoginTime[i]+int(12)
        if df.LoginTime[i] == int(24):
            df.LoginTime[i]=12
    
    elif df.Timing[i] == 'am' and df.LoginTime[i] == int(12):
        df.LoginTime[i]=0

df.head()
```

```python
df["Date_Time"] = df["LoginDate"]+" "+df["LoginTime"].astype(str)+":00"
df.head()
```

```python
del df['LoginDate']
del df['LoginTime']
del df['Timing']
```

```python
df["Date_Time"] = pd.to_datetime(df["Date_Time"])
```

```python
cnt = df.groupby(['Date_Time']).count()
cnt.columns=["Count"]
cnt.head()
```

```python
cnt.to_csv("rnn_data.csv")
```
